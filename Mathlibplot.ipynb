# Bar Plot

```python
import matplotlib.pyplot as plt
```

```python
x=["pthon","cpp","java","c"]
y=[85,70,60,56]
z=[20,30,40,50]
plt.xlabel("language" ,fontsize=20)
plt.ylabel("number",fontsize=20)
plt.title("wscube")
c=["y","b","m","g"]
plt.bar(x,y ,width=0.5,color="r",label="popularity")
plt.bar(x,z ,width=0.5,color="g",label="popularity1")
plt.legend()
plt.show()
```

```python
import numpy as np
x=["pthon","cpp","java","c"]
y=[95,70,60,56]
z=[80,30,40,50]

width=0.2
p=np.arange(len(x))
p1=[j+width for j in p]

plt.xlabel("language" ,fontsize=20)
plt.ylabel("number",fontsize=20)
plt.title("wscube",fontsize=15)

plt.bar(p,y ,width=0.2,color="r",label="popularity")
plt.bar(p1,z ,width=0.2,color="g",label="popularity1")

plt.xticks(p+width/2,x,rotation=20)
plt.legend()
plt.show()
```

# Scatter Plot

```python
day=[1,2,3,4,5,6,7]
no=[2,3,9,5,4,7,5]
colors=["g","r","y","r","b","g","r"]
sizes=[75,100,200,500,900,100,500]
plt.scatter(day,no,c=colors,s=sizes,alpha=0.2)
plt.title("scatter plot",fontsize=15)
plt.xlabel("Day",fontsize=16)
plt.ylabel("No",fontsize=15)
plt.show()
```
